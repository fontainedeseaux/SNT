### <u style="color:green;">2. Notion de balise :</u> ###
● HTML est un langage de description basé sur des <strong>balises</strong>. <br>
Elles servent à <strong>délimiter</strong> et <strong>décrire</strong> la signification de ce qu'elles contiennent.<br>
<br>
→ Par exemple, le mot <span style = "font-size:30px;">Bonjour</span> apparait ainsi, comme un titre car il est encadré par la paire balises <span style="font-family:Courier New;font-size: 100%;">&lt;h1&gt;...&lt;/h1&gt;</span> qui signifie : <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;*mets le contenu en tant que titre de niveau 1* (le plus grand niveau de titre possible).<br>
<br>
→ La phrase "Bienvenue sur mon site" est contenu dans la paire de balises <span style="font-family:Courier New;font-size: 100%;">&lt;p&gt; ... &lt;/p&gt;</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;Ce qui indique le contenu correspond à un <strong>p</strong>aragraphe.<br>
<br>
● Plus généralement, c’est grâce à des balises que l’on précise qu’un texte est un <u>titre</u>, un <u>paragraphe</u>, un <u>lien</u>, etc.<br>
Ces balises ont des noms <strong>spécifiques</strong>, selon les informations qu'elles définissent/délimitent (cf 3).<br>
<br>
● Une balise s'ouvre et se ferme ainsi : <br>
<span style="font-family:Courier New;font-size: 100%;">&lt;balise&gt; ... &lt;/balise&gt;</span><br>
<br>
● Les balises <strong>s'imbriquent dans l'ordre</strong>, comme les parenthèses (ou les crochets) en mathématiques :<br>
<span style="font-family:Courier New;font-size: 100%;">&lt;balise1&gt; ... &lt;balise2&gt; ... &lt;/balise2&gt; ...&lt;/balise1&gt;</span><br>
<br>
● Pour plus de clarté, on pourra utliser des <strong>indentations</strong> en cas de balises imbriquées :<br>
<span style="font-family:Courier New;font-size: 100%;">&lt;balise1&gt;</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">&lt;balise2&gt;</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">...</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">&lt;/balise2&gt;</span><br>
<span style="font-family:Courier New;font-size: 100%;">&lt;/balise1&gt;</span>
### <u style="color:green;">3. Structure d'un fichier html :</u> ###
Un document HTML comporte <strong>obligatoirement</strong> :<br>
● La balise <span style="font-family:Courier New;font-size: 100%;">&lt;!DOCTYPE html&gt;</span>.<br>
C'est une balise <strong>orpheline</strong> (il n'y a pas de paire). On dit qu'elle est <strong>auto-fermante</strong>.<br>
Son rôle est d'indiquer le <u>type de document</u>, c'est à dire qu'il s'agit d'un fichier codé en HTML5.<br>
<br>
● La paire de balises <span style="font-family:Courier New;font-size: 100%;">&lt;html&gt;&lt;/html&gt;</span> : l'élément <span style="font-family:Courier New;font-size: 100%;">&lt;html&gt;</span><br>
Grace à l'attribut <span style="font-family:Courier New;font-size: 100%;">lang</span> de cette balise, on peut apporter l'information d'une page rédigée en français.<br>
Tout le code de la page y est inclus. Ce code contient <strong>deux parties</strong>.<br>
<br>
● La paire de balises <span style="font-family:Courier New;font-size: 100%;">&lt;head&gt;&lt;/head&gt;</span> : l'élément <span style="font-family:Courier New;font-size: 100%;">&lt;head&gt;</span>.<br>
Il contient <strong>l’en-tête</strong> de la page. Ce contenu <strong>n'est pas affiché par le navigateur</strong> (à part le titre).<br>
On y retrouve :<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;→ le titre de la page : celui qui s’affiche <u>dans l’onglet du navigateur</u> (entre les balises <span style="font-family:Courier New;font-size: 100%;">&lt;title&gt;&lt;/title&gt;</span> ).<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;→  en général un lien vers le fichier CSS (cf)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;→  ainsi que des <strong>méta-données</strong> que nous n'abordons pas ici<br>
<br>
● la paire de balises <span style="font-family:Courier New;font-size: 100%;">&lt;body&gt;&lt;/body&gt;</span> : l'élément <span style="font-family:Courier New;font-size: 100%;">&lt;body&gt;</span><br>
<br><div style="text-align:center;"><span style="border: 2px solid;padding:6px;">Son contenu correspont à <strong>ce qui est réellement affiché</strong> par le navigateur.</span></div>

## <u style="color:red;">IV. Insertion d'images et de liens :</u> ##
### <u style="color:green;">1. Insertion d'une image dans une page :</u> ###
Pour afficher une image, on utilise la balise orpheline <span style="font-family:Courier New;font-size: 100%;">&lt;img&gt;</span><br>
Mais il est nécessaire de préciser <strong>où</strong> trouver cette image.<br>
La balise <span style="font-family:Courier New;font-size: 100%;">&lt;img&gt;</span> dispose donc d'un <strong>attribut</strong> <span style="font-family:Courier New;font-size: 100%;">src</span> (<strong>s</strong>ea<strong>rc</strong>h) qu'il convient de renseigner en précisant l'<strong>url</strong> de l'image à afficher :<br>
<br>
<span style="font-family:Courier New;font-size: 100%;">&lt;img src="url_du_lien_vers_l'image_a_afficher"&gt;</span><br>
<br>
Deux cas de figure se présentent alors :<br>
<br>
→ *Cas n° 1 :*<br>
L'image est <strong>physiquement présente</strong> dans le dossier <span style="font-family:Courier New;font-size: 100%;">images</span> de <span style="font-family:Courier New;font-size: 100%;">mon_site</span>.<br>
Insérons l'image nommée <span style="font-family:Courier New;font-size: 100%;">snt.png</span> dans la page <span style="font-family:Courier New;font-size: 100%;">index.html</span> :<br>
<span style="font-family:Courier New;font-size: 100%;">&lt;img src="images/snt.png"&gt;</span><br>
<br>
→ *Cas n° 2 :*<br>
L'image n'est pas physiquement présente dans un dossier, mais on dispose de son <span style="font-family:Courier New;font-size: 100%;">url</span> :<br>
https://isnpem.weebly.com/uploads/1/0/4/8/10485883/1537248_1.jpg<br>
(Qu'on obtient avec un *clic droit* sur l'image qui nous intéresse sur un site, puis *copier adresse*) <br>
<br>
Dans ce cas, on on remplace le nom de l'image par son <span style="font-family:Courier New;font-size: 100%;">url</span> dans l'attribut <span style="font-family:Courier New;font-size: 100%;">src</span>.<br>
<span style="font-family:Courier New;font-size: 100%;">&lt;img src="https://isnpem.weebly.com/uploads/1/0/4/8/10485883/1537248_1.jpg"&gt;</span><br>
Evidemment, elle ne sera affichable que si l'on est connecté à internet.
### <u style="color:green;">2. Insertion d'un lien cliquable :</u> ###
La balise <span style="font-family:Courier New;font-size: 100%;">&lt;a&gt;&lt;/a&gt;</span> est utilisée pour <strong>rendre cliquable son contenu</strong> afin créer un <u>lien</u> qui amène vers une page <strong>cible</strong>.<br>
Comme pour l'insertion d'une image, il s'agit de <strong>localiser</strong> cette page cible.<br>
Ce qui se fait en complétant de l'attribut <span style="font-family:Courier New;font-size: 100%;">href</span> : (<strong>h</strong>yperlink <strong>ref</strong>erence)<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• à l'aide d'un chemin relatif s'il s'agit d'une autre page de notre site<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• à l'aide d'une <span style="font-family:Courier New;font-size: 100%;">url</span> si c'est une page externe de notre site <br>
<br>
<span style="font-family:Courier New;font-size: 100%;">&lt;a href="url_du_lien_de_la_page_a_afficher"&gt; contenu cliquable &lt;/a&gt;</span><br>
<br>
<u>*Exemple :*</u><br>
● Dans notre page <span style="font-family:Courier New;font-size: 100%;">index.html</span>, créons un lien vers l'autre page de notre site : la page <span style="font-family:Courier New;font-size: 100%;">client_serveur.html</span>  :<br>
Puisque ces pages sont toutes deux dans le même dosier, on écrit :<br>
<span style="font-family:Courier New;font-size: 100%;">&lt;a href='client_serveur.html'> Cliquez ici &lt;/a> pour découvrir le modèle client-serveur...</span><br>

Ecrivez-y : <br>
<span style="font-family:Courier New;font-size: 100%;">p {</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">color : green;</span><br>
<span style="font-family:Courier New;font-size: 100%;">}</span><br>
<br>
Il s'agit d'un fichier "<strong>feuille de style</strong>" (<span style="font-family:Courier New;font-size: 100%;">=stylesheet</span>) que nous appliquons à la page <span style="font-family:Courier New;font-size: 100%;">client_serveur.html</span>.<br>
<br>
● Dans la page <span style="font-family:Courier New;font-size: 100%;">client_serveur.html</span>, insérons dans l'élément <span style="font-family:Courier New;font-size: 100%;">&lt;head&gt;</span> un <strong>lien</strong> (<span style="font-family:Courier New;font-size: 100%;">=link</span>) vers cette page de style.<br>
On utilise la balise orpheline <span style="font-family:Courier New;font-size: 100%;">&lt;link&gt;</span>, avec trois <strong>attributs</strong> <span style="font-family:Courier New;font-size: 100%;">href</span>, <span style="font-family:Courier New;font-size: 100%;">rel</span> et <span style="font-family:Courier New;font-size: 100%;">type</span> :<br>
On indique le <u>chemin</u> du fichier de style :<br>
<span style="font-family:Courier New;font-size: 100%;">&lt;link href="style.css" rel="stylesheet" type="text/css" &gt;</span><br>
<br>
● Relancez alors la page <span style="font-family:Courier New;font-size: 100%;">clients_serveur.html</span> : vous devriez constater que tous les paragraphes sont écrits en vert!<br>
Or, on peut avoir envie de mettre seulement <strong>une balise précise</strong> de cette couleur.<br>
Ou encore de créer un <strong>groupe de balises</strong> (=une <strong>class</strong>e) ayant un style particulier en commun.<br>
Il s'agit donc de pouvoir <strong>sélectionner</strong> un élément ou une famille d'élement html.<br>
C'est justement ce que permettent de faire les <strong>selecteurs CSS</strong>.
### <u style="color:green;">2. Sélectionner à l'aide d'un identifiant :</u> ###
Une règle Css telle que :<br>
<span style="font-family:Courier New;font-size: 100%;">p {</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">color : green;</span><br>
<span style="font-family:Courier New;font-size: 100%;">}</span><br>
concernera <strong>tous</strong> les éléments de type paragraphe.<br>
On peut ainsi facilement définir des comportements «par défaut» pour les éléments essentiels d'une page.<br>
<br>
● On peut désigner un élément <strong>unique</strong> dans une page avec l'attribut <span style="font-family:Courier New;font-size: 100%;">id</span>.<br>
Voici comment écrire en rouge le paragraphe commençant par "Grâce à son navigateur" :<br>
<br>
● <u>Dans le fichier <span style="font-family:Courier New;font-size: 100%;">client_serveur.html</span></u> :<br>
on ajoute un attribut <span style="font-family:Courier New;font-size: 100%;">id</span> à ce paragraphe. Par exemple : <span style="font-family:Courier New;font-size: 100%;">id = "role_navigateur"</span><br>
<span style="font-family:Courier New;font-size: 100%;">&lt;p id="role_navigateur"&gt; Grâce à son navigateur... &lt;/p&gt;</span><br>
<br>
● <u>Dans le fichier <span style="font-family:Courier New;font-size: 100%;">style.css</span></u>, on utilise un <span style="font-family:Courier New;font-size: 100%;">&#35;</span> devant l'identifiant :<br>
<span style="font-family:Courier New;font-size: 100%;">&#35;role_navigateur{</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">/* ATTENTION : ne pas oublier le &#35; devant le nom de l'identifiant */</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">color: blue;</span><br>
<span style="font-family:Courier New;font-size: 100%;">    }</span>
### <u style="color:green;">3. Les propriétés de couleur :</u> ###
De nombreux éléments d'une page web peuvent être mis en couleur :<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• la propriété <span style="font-family:Courier New;font-size: 100%;">color</span> modifie la couleur du texte<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• la propriété <span style="font-family:Courier New;font-size: 100%;">background-color</span> modifie la couleur d'arrière-plan<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• la propriété <span style="font-family:Courier New;font-size: 100%;">border-color</span> modifie la couleur des bordures<br>
<br>
Les valeurs prises par ces propriétés peuvent être :<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• une couleur nommée. On trouvera 500 couleurs nommées ici : http://www.profil-couleur.com/pp/326-noms-500-couleur.html<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;*Exemple :* <span style="font-family:Courier New;font-size: 100%;">color : aquamarine</span>        <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• ou une couleur représentée par un codage <span style="font-family:Courier New;font-size: 100%;">RGB</span>, à l'aide des 3 canaux <span style="font-family:Courier New;font-size: 100%;">RGB</span> : <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;*Exemple :* <span style="font-family:Courier New;font-size: 100%;">color : RGB(10, 153, 255)</span>.
### <u style="color:green;">4. Les Bordures :</u> ###
Chaque élément html est contenu dans une "boîte".<br>
Cette boîte possède des <strong>bordures</strong> modifiables grâce à trois propriétés :<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• <span style="font-family:Courier New;font-size: 100%;">border-width</span> pour l'épaisseur ;<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• <span style="font-family:Courier New;font-size: 100%;">border-style</span> pour la forme ;<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• <span style="font-family:Courier New;font-size: 100%;">border-color</span> pour la couleur.<br>
<br>
On peut aussi utiliser la propriété raccourci <span style="font-family:Courier New;font-size: 100%;">border</span> qui permet de préciser les trois propriétés précédentes en une seule déclaration. <br>
On peut tester les valeurs possibles pour ces propriétés sur cette page :<br>
https://developer.mozilla.org/fr/docs/Web/CSS/border-style<br>
<br>
<u>*Exemples :*</u><br>
<span style="font-family:Courier New;font-size: 100%;">p {</span>                             <br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">/* Mise en forme des paragraphes */</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">border-width : 6px;</span> <br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">border-style : dashed;</span> <br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">border-color : #0C18F4;</span> <br>
<span style="font-family:Courier New;font-size: 100%;">    }</span><br>
<br>
Raccourci possible en Css :<br>
<span style="font-family:Courier New;font-size: 100%;">p {</span> <br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">border : 6px dashed #0C18F4;</span> <br>
<span style="font-family:Courier New;font-size: 100%;">    }</span><br>

### <u style="color:green;">5. Dimensions :</u> ###
Nous venons de voir que chaque élément html est contenu dans une boîte. <br>
Pour comprendre le comportement des boîtes, il faut savoir ce que désigne chaque dimension utilisée.<br>
![alt text](pic/boite.png "Effet de bordures")<br>
Par défaut, les propriétés <span style="font-family:Courier New;font-size: 100%;">width</span> et <span style="font-family:Courier New;font-size: 100%;">height</span> désignent la largeur et la hauteur du <strong>contenu</strong> de la boîte.<br>
Pour avoir les dimensions de la boîte complète, il faut donc ajouter la valeur de :<br>
● <span style="font-family:Courier New;font-size: 100%;">padding</span> : dimensions entre le contenu et les bordures, qu'on peut donner en <span style="font-family:Courier New;font-size: 100%;">px</span><br>
https://developer.mozilla.org/fr/docs/Web/CSS/padding<br>
<br>
● <span style="font-family:Courier New;font-size: 100%;">border-width</span> : taille des bordures de la boîte<br>
https://developer.mozilla.org/fr/docs/Web/CSS/border-width<br>
<br>
● <span style="font-family:Courier New;font-size: 100%;">margin</span> : dimensions extérieures de la boîte.<br>
https://developer.mozilla.org/fr/docs/Web/CSS/margin<br>
<br>
Les propriétés précédentes peuvent être différenciées entre le haut (<span style="font-family:Courier New;font-size: 100%;">-top</span>), le bas (<span style="font-family:Courier New;font-size: 100%;">-bottom</span>), la gauche (<span style="font-family:Courier New;font-size: 100%;">-left</span>) et la droite (<span style="font-family:Courier New;font-size: 100%;">right</span>).<br>
On peut par exemple écrire : <span style="font-family:Courier New;font-size: 100%;">paddind-top=5px</span>.<br>
<br>
<u>*Exemples :*</u><br>
<span style="font-family:Courier New;font-size: 100%;">p {</span>                                     <br>
<span style="font-family:Courier New;font-size: 100%;">/* Mise en forme des paragraphes */</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">width : 400px ;</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">height : 100px ;</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">border : 6px double #C81E50;</span> <br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">padding : 10px ;</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">margin : 10px ;</span><br>
&nbsp;&nbsp;&nbsp;&nbsp;<span style="font-family:Courier New;font-size: 100%;">}</span><br>
